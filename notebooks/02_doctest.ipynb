## 1. Podstawy doctest

Doctest wykorzystuje sesje interpretera Pythona jako testy.

```python
def add(a, b):
    """
    Dodaje dwie liczby.
    
    >>> add(2, 3)
    5
    >>> add(-1, 1)
    0
    >>> add(0, 0)
    0
    """
    return a + b

def multiply(a, b):
    """
    MnoÅ¼y dwie liczby.
    
    >>> multiply(3, 4)
    12
    >>> multiply(-2, 3)
    -6
    >>> multiply(0, 100)
    0
    """
    return a * b

def factorial(n):
    """
    Oblicza silniÄ™ liczby n.
    
    >>> factorial(0)
    1
    >>> factorial(1)
    1
    >>> factorial(5)
    120
    >>> factorial(3)
    6
    """
    if n < 0:
        raise ValueError("n must be non-negative")
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
import doctest

# Uruchomienie doctest dla moduÅ‚u
print("ðŸ§ª Uruchamianie doctest...")

# Testowanie poszczegÃ³lnych funkcji
doctest.run_docstring_examples(add, globals(), verbose=True)
doctest.run_docstring_examples(multiply, globals(), verbose=True)
doctest.run_docstring_examples(factorial, globals(), verbose=True)

print("âœ… Podstawowe testy doctest zakoÅ„czone")
```
