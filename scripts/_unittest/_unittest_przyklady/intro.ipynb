## Podstawowa skÅ‚adnia

```python
assert warunek, "komunikat_bÅ‚Ä™du"
```

JeÅ›li `warunek` jest `False`, rzucany jest `AssertionError` z komunikatem.

````python
def divide(a, b):
    return a / b

# Proste testy z assert
assert divide(10, 5) == 2, "10/5 should equal 2"
assert divide(9, 3) == 3, "9/3 should equal 3"
print("âœ… Wszystkie testy przeszÅ‚y!")
````

## PrzykÅ‚ad z bÅ‚Ä™dem

````python
def square_sum(a, b):
    return (a + b) ** 2

# Testy ktÃ³re przejdÄ…
assert square_sum(1, 1) == 4, "(1+1)Â² should be 4"
assert square_sum(0, 0) == 0, "(0+0)Â² should be 0"
print("âœ… Pierwsze testy OK")

# Test ktÃ³ry nie przejdzie (odkomentuj Å¼eby zobaczyÄ‡ bÅ‚Ä…d)
assert square_sum(2, 3) == 20, "(2+3)Â² should be 25, not 20!"
````

## Asercje sÅ‚uÅ¼Ä… nie tylko do testowania!

Asercji uÅ¼ywa siÄ™ teÅ¼ czasamo do **walidacji** i **definiowania warunkÃ³w wstÄ™pnych** w kodzie produkcyjnym:

````python
def safe_divide(a, b):
    """Dzielenie z walidacjÄ…."""
    assert b != 0, "Cannot divide by zero"
    assert isinstance(a, (int, float)), "First argument must be a number"
    assert isinstance(b, (int, float)), "Second argument must be a number"
    return a / b

# UÅ¼ycie
print(safe_divide(10, 2))  # 5.0
````

````python
# Te wywoÅ‚ania rzucÄ… AssertionError (odkomentuj Å¼eby sprawdziÄ‡):
safe_divide(10, 0)        # Cannot divide by zero
````

````python
safe_divide("10", 2)      # First argument must be a number
````

## WiÄ™cej przykÅ‚adÃ³w testÃ³w

````python
def concatenate(str1, str2):
    return str1 + str2

# Testy funkcji concatenate
assert concatenate("hello", "world") == "helloworld"
assert concatenate("", "") == ""
assert concatenate("a", "b") == "ab"
print("âœ… Testy concatenate OK")

# Test listy
numbers = [1, 2, 3, 4, 5]
assert len(numbers) == 5
assert max(numbers) == 5
assert min(numbers) == 1
assert sum(numbers) == 15
print("âœ… Testy listy OK")
````

## âš ï¸ WaÅ¼ne ostrzeÅ¼enie: Flaga -O

**Asercje mogÄ… byÄ‡ wyÅ‚Ä…czone!** Gdy Python uruchamiany jest z flagÄ… `-O` (optymalizacja), wszystkie asercje sÄ… **ignorowane**.

**Flaga -O**: Usuwa asercje + docstringi, tworzy `.pyo` pliki (zoptymalizowany bytecode zamiast standardowego `.pyc`). UÅ¼ywana w produkcji dla wydajnoÅ›ci.

````python
# SprawdÅº czy asercje sÄ… wÅ‚Ä…czone
print(f"__debug__ = {__debug__}")

if __debug__:
    print("âœ… Asercje sÄ… wÅ‚Ä…czone")
else:
    print("âš ï¸ Asercje sÄ… WYÅÄ„CZONE (python -O)")
````

### Skoro asercje moÅ¼na wyÅ‚Ä…czyÄ‡...

**Assert = logika deweloperska** (zaÅ‚oÅ¼enia programisty):
- "Ten parametr zawsze bÄ™dzie dodatni"
- "Ta lista nigdy nie bÄ™dzie pusta"

**Logika biznesowa = walidacja uÅ¼ytkownika:**
```python

# âœ… DOBRE - zawsze dziaÅ‚a

    raise ValueError("Wiek musi byÄ‡ dodatni")
```

Assert to narzÄ™dzie **debugowania i testowania**, nie czÄ™Å›Ä‡ funkcjonalnoÅ›ci aplikacji.

````python
# PrzykÅ‚ady bardziej zÅ‚oÅ¼onych asercji
data = [1, 2, 3, 4, 5]

# SprawdÅº czy wszystkie elementy sÄ… pozytywne
assert all(x > 0 for x in data), "All numbers should be positive"

# SprawdÅº czy przynajmniej jeden element jest parzysty
assert any(x % 2 == 0 for x in data), "At least one number should be even"

# SprawdÅº typ
assert isinstance(data, list), "Data should be a list"

print("âœ… ZÅ‚oÅ¼one testy OK")
````
